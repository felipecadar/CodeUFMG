```python
import matplotlib.pyplot as plt
%matplotlib inline

!wget -nc -O "RomeoAndJuliet.txt" http://www.gutenberg.org/files/1112/1112.txt
```

```python
filename = "RomeoAndJuliet.txt"

f = open(filename)
lines = f.readlines()
```

Abra o arquivo e tente plotar algumas informaçoes

1. Plot de barra das 10 palavras mais usadas

```python
h = {}
for l in lines:
    l = l.replace(",", "")
    l = l.replace(".", "")
    l = l.replace("*", "")
    l = l.replace("\n", "")
    for p in l.split(" "):
        p = p.lower()
        if p in h:
            h[p] += 1
        else:
            h[p] = 1

top_10 = sorted(h.items(), key=lambda x: x[1])[::-1][1:11]
values = [i[1] for i in top_10]
names = [i[0] for i in top_10]

plt.figure(figsize=(9, 9))
plt.bar(names, values)
plt.show()
```

2. Plot de barra com a quantidade de vezes que o nome dos personagens principais foram citados

```python
h = {"romeo": 0, "juliet": 0}
for l in lines:
    l = l.replace(",", "")
    l = l.replace(".", "")
    l = l.replace("*", "")
    l = l.replace("\n", "")
    for p in l.split(" "):
        p = p.lower()
        if p in h:
            h[p] += 1

values = [i[1] for i in h.items()]
names = [i[0] for i in h.items()]

plt.figure(figsize=(9, 9))
plt.bar(names, values)
plt.show()
```

3. Scatter plot relacionando a quantidade de letras com a quantidade de palavras em cada linha do texto 

```python
x = []
y = []
for l in lines:
    palavras = len(l.split(" "))
    letras = len(l.replace(" ", ""))
    x.append(palavras)
    y.append(letras)
    

plt.figure(figsize=(9, 9))
plt.xlabel("# Palavras")
plt.ylabel("# Letras")
plt.scatter(x, y, alpha=0.5)
plt.show()
```
